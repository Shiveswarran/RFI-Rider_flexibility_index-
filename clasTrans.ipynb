```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.patches as mpatches
import seaborn as sns
from scipy import stats
from scipy.stats import kruskal, mannwhitneyu, chi2_contingency
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore')

plt.rcParams.update({
    'figure.dpi': 130, 'font.family': 'DejaVu Sans',
    'axes.spines.top': False, 'axes.spines.right': False,
    'axes.titlesize': 12, 'axes.titleweight': 'bold',
    'axes.labelsize': 10, 'xtick.labelsize': 9, 'ytick.labelsize': 9,
    'legend.fontsize': 9, 'figure.facecolor': 'white',
    'axes.facecolor': '#f8f8f6', 'axes.grid': True,
    'grid.color': 'white', 'grid.linewidth': 0.8,
})

print('✓ Imports ready')
```

```python
def parse_dt(series):
    """Parse datetime strings to pandas Timestamp. Handles M/D/YY H:MM and ISO."""
    return pd.to_datetime(series, errors='coerce', infer_datetime_format=True)

def parse_mins(series):
    """Parse time-only or datetime strings to minutes since midnight."""
    def _one(v):
        if pd.isna(v) or str(v).strip() in ('', 'nan', 'NaT'): return np.nan
        s = str(v).strip()
        if ' ' in s:
            try:
                dt = pd.to_datetime(s, infer_datetime_format=True)
                return dt.hour*60 + dt.minute
            except: return np.nan
        try:
            parts = s.split(':')
            return int(parts[0])*60 + int(parts[1])
        except: return np.nan
    return series.apply(_one)

# Full datetime columns
eco['Trip Date']    = parse_dt(eco['Trip Date'])
eco['Order Time']   = parse_dt(eco['Order Time'])
eco['Cancel Time']  = parse_dt(eco['Cancel Time'])

# Time-of-day columns (minutes since midnight)
time_cols = [
    ('Promised Pick-up Time',          'promised_pu_min'),
    ('Requested Pick-up Time',         'req_pu_min'),
    ('Earliest Pick-up',               'earliest_pu_min'),
    ('Latest Pick-up Time',            'latest_pu_min'),
    ('Scheduled Pick-up Arrival Time', 'sched_pu_min'),
    ('Reported Pick-up Arrival Time',  'rep_pu_min'),
    ('Reported Pick-up Departure Time','rep_pu_dep_min'),
    ('Requested Drop-off Time',        'req_do_min'),
    ('Scheduled Drop-Off Arrival Time','sched_do_min'),
    ('Reported Drop-Off Arrival Time', 'rep_do_min'),
    ('Earliest Drop-off Time',         'earliest_do_min'),
    ('Latest Drop-off Time',           'latest_do_min'),
]
for raw, col in time_cols:
    if raw in eco.columns:
        eco[col] = parse_mins(eco[raw])

print('── Parse success rates ──────────────────────────────────────')
for raw, col in time_cols:
    if col in eco.columns:
        n = eco[col].notna().sum()
        print(f'  {raw:<40}: {n:>7,}  ({n/len(eco)*100:.1f}%)')

# Calendar features
eco['Trip Date'] = eco['Trip Date'].fillna(
    pd.to_datetime(eco['Scheduled Pick-up Arrival Time'], errors='coerce').dt.normalize()
)
eco['dow']       = eco['Trip Date'].dt.day_name()
eco['month']     = eco['Trip Date'].dt.to_period('M').astype(str)
eco['pu_hour']   = (eco['earliest_pu_min'] // 60).astype('Int64')
eco['time_bin']  = pd.cut(eco['earliest_pu_min'],
                            bins=[0,360,480,600,720,840,960,1440],
                            labels=['<6am','6-8am','8-10am','10am-12pm',
                                    '12-2pm','2-4pm','4pm+'])

# Auto-detect status and purpose values
COMPLETED_LABEL  = eco['Trip Status'].value_counts().index[0]
PURPOSE_ORDER    = eco['Purpose'].dropna().value_counts().index.tolist()
_colors = ['#1D9E75','#7F77DD','#EF9F27','#D4537E','#378ADD',
            '#E24B4A','#639922','#BA7517']
PALETTE = {p: _colors[i % len(_colors)] for i, p in enumerate(PURPOSE_ORDER)}

print(f'\nCOMPLETED_LABEL : {repr(COMPLETED_LABEL)}')
print(f'PURPOSE_ORDER   : {PURPOSE_ORDER}')
print('✓ Parsing complete')
```

```python
# ── SIGNAL 1: Negotiation Window (direct scheduler flexibility signal) ─
if 'Negotiation Window Minus' in eco.columns:
    eco['neg_window_minus'] = pd.to_numeric(eco['Negotiation Window Minus'], errors='coerce')
    eco['neg_window_plus']  = pd.to_numeric(eco['Negotiation Window Plus'],  errors='coerce')
    eco['neg_window_total'] = eco['neg_window_minus'].fillna(0) + eco['neg_window_plus'].fillna(0)

# ── SIGNAL 2: Promised vs Requested shift ───────────────────────
eco['promised_shift'] = eco['promised_pu_min'] - eco['req_pu_min']  # positive = pushed later
eco['promised_shift_abs'] = eco['promised_shift'].abs()

# ── SIGNAL 3: Pickup Window (earliest/latest difference) ────────
eco['pickup_window'] = (eco['latest_pu_min'] - eco['earliest_pu_min']).clip(lower=0)
eco['dropoff_window']= (eco['latest_do_min'] - eco['earliest_do_min']).clip(lower=0)

# ── SIGNAL 4: Booking lead time (days) ──────────────────────────
eco['booking_lead_days'] = (
    (eco['Trip Date'] - eco['Order Time'].dt.normalize()).dt.days
).clip(lower=0)
eco['is_subscription'] = eco['Subscription Number'].notna().astype(int)

# ── SIGNAL 5: OTP absorption ────────────────────────────────────
eco['otp_min'] = pd.to_numeric(eco['Pick-up OTP (minutes)'], errors='coerce')

# ── SIGNAL 6: Actual vs scheduled deviation ─────────────────────
eco['sched_deviation'] = eco['rep_pu_min'] - eco['sched_pu_min']  # +ve = late
eco['sched_deviation_abs'] = eco['sched_deviation'].abs()

# ── SIGNAL 7: Actual vs estimated efficiency ────────────────────
eco['dist_actual']    = pd.to_numeric(eco.get('Actual Trip Distance'), errors='coerce')
eco['dist_estimated'] = pd.to_numeric(eco.get('Direct Estimated Distance'), errors='coerce')
eco['dur_actual']     = pd.to_numeric(eco.get('Actual Trip Duration'), errors='coerce')
eco['dur_estimated']  = pd.to_numeric(eco.get('Direct Estimated Duration'), errors='coerce')
eco['dist_ratio'] = (eco['dist_actual'] / eco['dist_estimated'].replace(0, np.nan)).clip(0.5, 5)
eco['dur_ratio']  = (eco['dur_actual']  / eco['dur_estimated'].replace(0, np.nan)).clip(0.5, 5)

# ── SIGNAL 8: Loading/unloading (mobility complexity) ───────────
eco['loading_min']   = pd.to_numeric(eco.get('Loading Duration'),   errors='coerce')
eco['unloading_min'] = pd.to_numeric(eco.get('Unloading Duration'), errors='coerce')
eco['mobility_time'] = eco['loading_min'].fillna(0) + eco['unloading_min'].fillna(0)

# ── Cancel lead time ────────────────────────────────────────────
eco['cancel_lead_min'] = (
    (eco['Cancel Time'] - eco['Trip Date']).dt.total_seconds() / 60
).clip(-1440, 4320)  # -24h to +3days

# ── Subset: completed trips ──────────────────────────────────────
eco_c = eco[eco['Trip Status'] == COMPLETED_LABEL].copy()

print('── Signal availability (all trips) ─────────────────────────')
signals = [
    ('Negotiation window',   'neg_window_total'),
    ('Promised shift',       'promised_shift'),
    ('Pickup window',        'pickup_window'),
    ('Booking lead days',    'booking_lead_days'),
    ('OTP',                  'otp_min'),
    ('Sched deviation',      'sched_deviation'),
    ('Dist ratio',           'dist_ratio'),
    ('Cancel lead',          'cancel_lead_min'),
    ('Mobility time',        'mobility_time'),
    ('Subscription flag',    'is_subscription'),
]
for label, col in signals:
    if col in eco.columns:
        n = eco[col].notna().sum()
        print(f'  {label:<22}: {n:>7,}  ({n/len(eco)*100:.1f}%)')

print(f'\nCompleted trips : {len(eco_c):,}')
```

```python
fig, axes = plt.subplots(2, 3, figsize=(17, 10), constrained_layout=True)
fig.suptitle('Negotiation Window — Scheduler-Assumed Flexibility', fontsize=14, fontweight='bold')
avail_p = [p for p in PURPOSE_ORDER if p in eco_c['Purpose'].unique()]
dow_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

# 1. Overall distribution of total negotiation window
ax = axes[0, 0]
nw = eco_c['neg_window_total'].dropna()
ax.hist(nw, bins=50, color='#7F77DD', edgecolor='white', linewidth=0.3)
for q, c, l in zip([0.25,0.5,0.75],['#E24B4A','#333','#1D9E75'],['Q1','Median','Q3']):
    v = nw.quantile(q)
    ax.axvline(v, color=c, lw=1.6, ls='--', label=f'{l}={v:.0f}min')
ax.set_title('Total negotiation window (all trips)')
ax.set_xlabel('Minus + Plus (minutes)'); ax.set_ylabel('Trips'); ax.legend(frameon=False)

# 2. Minus vs Plus asymmetry by purpose
ax = axes[0, 1]
minus_by_p = eco_c.groupby('Purpose')['neg_window_minus'].mean().reindex(avail_p)
plus_by_p  = eco_c.groupby('Purpose')['neg_window_plus'].mean().reindex(avail_p)
x = np.arange(len(avail_p)); w = 0.35
ax.bar(x - w/2, minus_by_p.fillna(0), w, color='#E24B4A',
       label='Neg window minus (early)', edgecolor='white')
ax.bar(x + w/2, plus_by_p.fillna(0),  w, color='#1D9E75',
       label='Neg window plus (late)',  edgecolor='white')
ax.set_xticks(x); ax.set_xticklabels(avail_p, rotation=25, ha='right')
ax.set_title('Negotiation window asymmetry by purpose\n(how early vs late the system can push the trip)')
ax.set_ylabel('Mean minutes'); ax.legend(frameon=False)

# 3. Neg window vs promised shift (do wider windows mean bigger shifts?)
ax = axes[0, 2]
valid = eco_c[eco_c['neg_window_total'].notna() & eco_c['promised_shift_abs'].notna()]
if len(valid) > 3000: valid = valid.sample(3000, random_state=42)
for p in avail_p:
    sub = valid[valid['Purpose']==p]
    if len(sub) > 10:
        ax.scatter(sub['neg_window_total'], sub['promised_shift_abs'],
                   color=PALETTE.get(p,'#888'), alpha=0.3, s=10,
                   label=p, edgecolors='none')
v = valid[['neg_window_total','promised_shift_abs']].dropna()
if len(v) > 5:
    m, b, r, _, _ = stats.linregress(v['neg_window_total'], v['promised_shift_abs'])
    xs = np.linspace(v['neg_window_total'].min(), v['neg_window_total'].max(), 100)
    ax.plot(xs, m*xs+b, 'k--', lw=1.5, label=f'r={r:.2f}')
ax.set_title('Negotiation window vs promised shift\n(wider window → bigger trip movement?)')
ax.set_xlabel('Negotiation window (min)'); ax.set_ylabel('|Promised − Requested| (min)')
ax.legend(frameon=False, markerscale=2, fontsize=7.5)

# 4. Neg window by subscription vs ad-hoc
ax = axes[1, 0]
sub_vals = eco_c[eco_c['is_subscription']==1]['neg_window_total'].dropna()
adhoc_vals= eco_c[eco_c['is_subscription']==0]['neg_window_total'].dropna()
ax.hist(sub_vals,   bins=40, alpha=0.6, color='#E24B4A', density=True,
        label=f'Subscription (n={len(sub_vals):,}, μ={sub_vals.mean():.0f}min)', edgecolor='none')
ax.hist(adhoc_vals, bins=40, alpha=0.6, color='#1D9E75', density=True,
        label=f'Ad-hoc (n={len(adhoc_vals):,}, μ={adhoc_vals.mean():.0f}min)', edgecolor='none')
if len(sub_vals)>1 and len(adhoc_vals)>1:
    U, p_u = mannwhitneyu(sub_vals, adhoc_vals, alternative='two-sided')
    ax.text(0.97, 0.97, f'Mann-Whitney p={p_u:.2e}',
            transform=ax.transAxes, ha='right', va='top', fontsize=8, color='gray')
ax.set_title('Negotiation window: subscription vs ad-hoc\n(hypothesis: subscriptions = narrower window)')
ax.set_xlabel('Negotiation window (min)'); ax.set_ylabel('Density'); ax.legend(frameon=False)

# 5. Neg window by time of day
ax = axes[1, 1]
tbin_nw = eco_c.groupby('time_bin', observed=True)['neg_window_total'].agg(['mean','sem']).dropna()
ax.barh(tbin_nw.index, tbin_nw['mean'],
        xerr=tbin_nw['sem'].fillna(0),
        color='#7F77DD', edgecolor='white',
        error_kw={'linewidth':1,'color':'#555'})
ax.set_title('Mean negotiation window by time bin\n(peak hours may get smaller windows)')
ax.set_xlabel('Mean negotiation window (min)')

# 6. Neg window heatmap: purpose × day
ax = axes[1, 2]
pivot_nw = eco_c.groupby(['Purpose','dow'])['neg_window_total'].mean().unstack()
pivot_nw = pivot_nw.reindex(avail_p).reindex(columns=dow_order, fill_value=np.nan)
if not pivot_nw.empty:
    sns.heatmap(pivot_nw, ax=ax, cmap='YlGnBu', annot=True, fmt='.0f',
                linewidths=0.4, cbar_kws={'shrink':0.8,'label':'mean min'},
                annot_kws={'size':7.5}, mask=pivot_nw.isna())
ax.set_title('Mean neg window: purpose × day')
ax.set_xticklabels([d[:3] for d in dow_order], rotation=0, fontsize=8)
ax.set_ylabel('')

plt.savefig('ec_fig1_neg_window.png', bbox_inches='tight', dpi=150)
plt.show()

print('\n── Negotiation window summary by purpose ───────────────────')
display(eco_c.groupby('Purpose')['neg_window_total']
        .describe(percentiles=[0.25,0.5,0.75]).round(1).reindex(avail_p))
```

```python
fig, axes = plt.subplots(2, 3, figsize=(17, 10), constrained_layout=True)
fig.suptitle('Promised vs Requested Shift — Rider Accepted Displacement', fontsize=14, fontweight='bold')

# 1. Overall shift distribution
ax = axes[0, 0]
shift = eco_c['promised_shift'].dropna().clip(-120, 120)
ax.hist(shift, bins=60, color='#EF9F27', edgecolor='white', linewidth=0.3)
ax.axvline(0, color='gray', lw=1.2, ls='--', alpha=0.7, label='No shift')
ax.axvline(shift.mean(), color='#E24B4A', lw=1.8, ls='--', label=f'Mean={shift.mean():.1f}min')
pct_positive = (eco_c['promised_shift'].dropna() > 0).mean() * 100
pct_negative = (eco_c['promised_shift'].dropna() < 0).mean() * 100
ax.text(0.97, 0.97, f'Pushed later: {pct_positive:.0f}%\nPushed earlier: {pct_negative:.0f}%',
        transform=ax.transAxes, ha='right', va='top', fontsize=8.5)
ax.set_title('Promised − Requested pick-up (min)\n+ve = pushed later, −ve = earlier')
ax.set_xlabel('Shift (minutes)'); ax.set_ylabel('Trips'); ax.legend(frameon=False)

# 2. Shift by purpose (boxplot)
ax = axes[0, 1]
bp = ax.boxplot(
    [eco_c[eco_c['Purpose']==p]['promised_shift'].dropna().clip(-120,120).values
     for p in avail_p],
    labels=avail_p, patch_artist=True, showfliers=False,
    medianprops={'color':'white','linewidth':2}
)
for patch, p in zip(bp['boxes'], avail_p):
    patch.set_facecolor(PALETTE.get(p,'#888')); patch.set_alpha(0.85)
ax.axhline(0, color='gray', lw=0.8, ls='--', alpha=0.5)
ax.set_title('Promised shift by purpose')
ax.set_ylabel('Shift (min, clipped ±120)'); ax.set_xticklabels(avail_p, rotation=25, ha='right')
if len(avail_p)>=2:
    groups = [eco_c[eco_c['Purpose']==p]['promised_shift'].dropna() for p in avail_p]
    groups = [g for g in groups if len(g)>5]
    if len(groups)>=2:
        H, pv = kruskal(*groups)
        ax.text(0.97,0.97,f'Kruskal-Wallis p={pv:.2e}',
                transform=ax.transAxes, ha='right', va='top', fontsize=8, color='gray')

# 3. Absolute shift by time bin
ax = axes[0, 2]
shift_tbin = eco_c.groupby('time_bin', observed=True)['promised_shift_abs'].agg(['mean','sem']).dropna()
ax.barh(shift_tbin.index, shift_tbin['mean'],
        xerr=shift_tbin['sem'].fillna(0),
        color='#EF9F27', edgecolor='white', error_kw={'linewidth':1,'color':'#555'})
ax.set_title('Mean |shift| by pickup time bin\n(are peak hours pushed more?)')
ax.set_xlabel('|Promised − Requested| (min)')

# 4. Rider-level: mean absolute shift distribution
ax = axes[1, 0]
rider_shift = eco_c.groupby('Customer Number')['promised_shift_abs'].mean().dropna()
ax.hist(rider_shift, bins=40, color='#EF9F27', edgecolor='white', linewidth=0.4)
ax.axvline(rider_shift.mean(), color='#E24B4A', lw=1.8, ls='--',
           label=f'Mean={rider_shift.mean():.1f}min')
ax.set_title('Per-rider mean absolute shift\n(consistently shifted = flexible)')
ax.set_xlabel('Mean |shift| (min)'); ax.set_ylabel('Riders'); ax.legend(frameon=False)

# 5. Subscription vs ad-hoc shift
ax = axes[1, 1]
for flag, label, col in [(1,'Subscription','#E24B4A'),(0,'Ad-hoc','#1D9E75')]:
    vals = eco_c[eco_c['is_subscription']==flag]['promised_shift_abs'].dropna()
    if len(vals)>0:
        ax.hist(vals.clip(upper=120), bins=35, alpha=0.55, color=col,
                label=f'{label} μ={vals.mean():.0f}min', density=True, edgecolor='none')
ax.set_title('Shift magnitude: subscription vs ad-hoc')
ax.set_xlabel('|Promised − Requested| (min)'); ax.set_ylabel('Density'); ax.legend(frameon=False)

# 6. Shift direction × Cancel Type (cancelled trips)
ax = axes[1, 2]
eco_can = eco[eco['Cancel Type'].notna() & eco['promised_shift'].notna()].copy()
top_cancel_types = eco_can['Cancel Type'].value_counts().head(6).index.tolist()
bp2 = ax.boxplot(
    [eco_can[eco_can['Cancel Type']==c]['promised_shift'].clip(-120,120).values
     for c in top_cancel_types],
    labels=top_cancel_types, patch_artist=True, showfliers=False,
    medianprops={'color':'white','linewidth':2}
)
_can_colors = ['#1D9E75','#7F77DD','#EF9F27','#E24B4A','#D4537E','#378ADD']
for patch, col in zip(bp2['boxes'], _can_colors):
    patch.set_facecolor(col); patch.set_alpha(0.85)
ax.axhline(0, color='gray', lw=0.8, ls='--', alpha=0.5)
ax.set_title('Promised shift for cancelled trips\n(were they cancelled because shifted?)')
ax.set_ylabel('Promised shift (min)'); ax.set_xticklabels(top_cancel_types, rotation=30, ha='right')

plt.savefig('ec_fig2_promised_shift.png', bbox_inches='tight', dpi=150)
plt.show()
```
